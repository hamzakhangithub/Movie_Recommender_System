```python
selected_features = ['genres','keywords','tagline','cast','director']
```

```python
combined_features = df['genres']+' '+df['keywords']+' '+df['tagline']+' '+df['cast']+' '+df['director']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cv = TfidfVectorizer()
```

```python
feature_vectors = cv.fit_transform(combined_features)
```

```python
similarity =cosine_similarity(feature_vectors)
```

```python
print(similarity.shape)
```

```python
import difflib
movie_name = input("Enter a movie name: ")

list_of_all_movies = df['title'].tolist()
#print(list_of_all_movies)

close_match = difflib.get_close_matches(movie_name,list_of_all_movies)
#print(close_match)

closest_match = close_match[0]

movie_index = df[df.original_title == closest_match]['index'].values[0]
#print(movie_index)

similarity_score = list(enumerate(similarity[movie_index]))

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)
```

```python
#print the movie names based on index

print("Recommended Movies: \n")

i = 1

for movies in sorted_similar_movies:
  index = movies[0]
  title_from_index = df[df.index == index]['original_title'].values[0]
  if i<30:
    print(i,'.',title_from_index)
    i+=1
```
